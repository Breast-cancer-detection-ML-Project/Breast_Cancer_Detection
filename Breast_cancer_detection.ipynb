## Data preprocessing 

```python
X = df.drop(['diagnosis'], axis = 1)
X.head(10)
```

```python
y = df['diagnosis'].values
m=len(y)
y=y.reshape(m,1)
```

## Split data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
print("x train: ",X_train.shape)
print("x test: ",X_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_sc,y_train)
```

```python
y_pred_lr = lr.predict(X_test_sc)
```

```python
y_pred_lr
```

```python
print(lr.score(X_test, y_pred_lr))
print(lr.score(X_test,y_test))
```

## Draw confusion matrix

```python
plt.figure(figsize=(15,8))
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_lr)
plt.title('Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
print('Confusion matrix : \n',confusion_matrix(y_test, y_pred_lr),'\n')
```

# create a function for models

```python
def models(x_train,y_train):
    # Logistic regression
    from sklearn.linear_model import LogisticRegression
    logis = LogisticRegression(random_state=0)
    logis.fit(x_train,y_train)

    #Decision Tree classifier
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(x_train,y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(x_train,y_train)

    # print the model accuracy of the training data
    print('[0]Logistic Regression Training Accuracy',logis.score(x_train,y_train))
    print('[1]Decision tree Classifier Training Accuracy', tree.score(x_train, y_train))
    print('[2]Random Forest Classifier Training  Accuracy', forest.score(x_train, y_train))
    return logis,tree,forest
```

```python
#getting all of the models
model = models(X_train_sc,y_train)

#test model accuracy on test data on confusion martix
from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    print('Model',i)
    cm = confusion_matrix(y_test, model[i].predict(X_test_sc))
    print(cm)
    plt.title('Confusion Matrix', fontsize = 15)
    sns.heatmap(cm, annot = True)
    plt.show()
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[0][1]
    FN = cm[1][0]
    print('testing accuracy=', (TP + TN) / (TP + TN + FP + FN))

#another way to get metrics of the models
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print('Model',i)
    print(classification_report(y_test, model[0].predict(X_test_sc)))
    print(accuracy_score(y_test, model[0].predict(X_test_sc)))


# print the prediction of rando forest classifier Model
pred = model[2].predict(X_test_sc)
print(pred)
print()
print(y_test)
print("Shivam rajput Gla university Mathura.")
```
